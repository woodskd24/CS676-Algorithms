```python
df1['Age-adjusted Death Rate'] = df1['Age-adjusted Death Rate'].astype(int)
df1['Region'] = df1['Region'].astype(int)
```

```python
import pandas as pd
import numpy as np

def pca_dataframe(df, n_components):

    X_centered = df1 - df1.mean()

    cov_matrix = np.cov(X_centered, rowvar=False)

    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]

    components = sorted_eigenvectors[:, :n_components]

    X_transformed = np.dot(X_centered, components)

    return X_transformed, components
```

```python
# Perform PCA on the DataFrame
#df2 = df1.drop('Region', axis=1)
df = df1.drop('Year', axis=1)
X_transformed, components = pca_dataframe(df, n_components=3)

# Print the transformed data and principal components
print(X_transformed)
print(components)
```
